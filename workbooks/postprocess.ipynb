```python
import pandas as pd

# Read the JSON file into a dataframe
df = pd.read_json('../companies_classifications.json')

# Display the dataframe
df
```

```python
len(df["classification"].unique())
df["classification"].unique()
```

```python
filters =  ['Technology/Software', 'Neuroscience','Technology Consulting', 'Technology', "Energy", "Software", "Cybersecurity", "Logistics", "Analytics", "Information Technology"]
filtered = df[df['classification'].isin(filters)]
```

```python
filtered.to_csv('filtered_data.csv', index=False)
```

```python
len(df[df["classification"] == "Technology"])
```

```python
column_counts = df['classification'].value_counts()
filtered_columns = column_counts[column_counts > 2]
filtered_columns.index = filtered_columns.index.str.replace(' ', '\n')
filtered_columns.index = filtered_columns.index.str.replace('Development', 'Dev')

import matplotlib.pyplot as plt

# Plot the filtered_columns dataframe
plt.figure(figsize=(16, 6))
filtered_columns.plot(kind='bar')

# Rotate the x-axis labels by 45 degrees
plt.xticks(rotation=40)
plt.ylabel("Count")
plt.title("Label Counts")
# Display the plot
plt.show()
```
